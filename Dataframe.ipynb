```python
# import packages                                                           
import numpy as np                                                                    
import pandas as pd
```

```python
# import original datesets
df_edges = pd.read_table('cit-HepTh.txt', names=['from', 'to'])
df_dates = pd.read_table('cit-HepTh-dates.txt', names=['id', 'date'])
```

```python
# set datetime format
df_dates["date"] = pd.to_datetime(df_dates["date"])
from datetime import datetime
```

```python
# merge two datasets
df_join1 = pd.merge(df_edges, df_dates, left_on='from', right_on='id').drop(columns='id')
df_join2 = pd.merge(df_join1, df_dates, left_on='to', right_on='id').drop(columns='id')
df_join2
```

```python
df_join2 = df_join2.sort_values(['date_y','to'])
df_join2000 = df_join2[df_join2['date_y'] > datetime(2000,1,1)] # pick the data after 2000-01-01
df_join2000
```

```python
df_join2000 = df_join2000.drop_duplicates(['from', 'to']) # drop duplicated data
df_join2000 = df_join2000[df_join2000['date_x'] > df_join2000['date_y']] # drop disordered data
df_join2000
```

```python
df_from = df_join2000.loc[:,['from','date_x']]
df_fromnew = df_from.rename(columns={'from': 'id', 'date_x': 'date'})

df_to = df_join2000.loc[:,['to','date_y']]
df_tonew = df_to.rename(columns={'to': 'id', 'date_y': 'date'})

df_netdates = pd.concat([df_tonew, df_fromnew],ignore_index=True).sort_values(['date','id']).drop_duplicates(['id']).reset_index()
df_netdates
```

```python
df_netdates['new_id']=range(len(df_netdates))
df_netdates
```

```python
df_join2000a = pd.merge(df_join2000, df_netdates, left_on='from', right_on='id').drop(columns='id').drop(columns='date').drop(columns='index')
df_join2000b = pd.merge(df_join2000a, df_netdates, left_on='to', right_on='id').drop(columns='id').drop(columns='date').drop(columns='index')
df_formatted = df_join2000b.sort_values(['new_id_y','date_y','to']).reset_index().drop(columns='index')
df_formatted
```

```python
df_formatted.to_csv("df_formatted.csv")
```
